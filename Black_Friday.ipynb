# Problem Statement

A retail company "ABC Private Limited" wanta to understand the customer purchases behaviour (specifically,purchased amount)
against various products of different categories. They Have shared piurchased summary of various products from last month.
The data set also contain customer demographics(age,gender,marital status, city_type,stay_in_current_city),product details(product_id and product_category)

```python
# Importing the dataset
df_train=pd.read_csv('train.csv')
df_train.head()
```

```python
# importing the test data
df_test=pd.read_csv("test.csv")
df_test.head()
```

```python
# merge the both dataset that is Train and Test data
df=df_train.append(df_test)
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.drop('User_ID',axis=1,inplace=True)
df.head(3)
```

```python
# Now coverting the Gender (Categorical features) into numerical features with map or replace parameter
df['Gender']=df['Gender'].map({'F':0,"M":1})
df.head()
```

```python
# Handle categorical features Age:
df['Age'].unique()
```

```python
#change the categorical feature(Age) into numerical feature:
df['Age']=df['Age'].map({'0-17':1,'18-25':2,'26-35':3,'36-45':4,'46-50':5,'51-55':6,'55+':7})
```

```python
df.head()
```

```python
df_city=pd.get_dummies(df['City_Category'],drop_first=True)
df_city.head()
```

```python
df.head()
```

```python
df=pd.concat([df,df_city],axis=1)
df.head()
```

```python
# drop the City_Category feature as we don't required it and we have make it in a seperate feature as B C
df.drop("City_Category",axis=1,inplace=True)
df.head()
```

### Focus on replacing the missing value

```python
df['Product_Category_2'].unique()  #it is a discrete variable
```

```python
print(df['Product_Category_2'].value_counts())
```

```python
df['Product_Category_2'].mode()[0]
```

```python
# the best way to replace the missing (discrete variable) value is mode:
df['Product_Category_2']=df['Product_Category_2'].fillna(df['Product_Category_2'].mode()[0])
```

```python
df['Product_Category_2'].isnull().sum()
```

```python
#df
```

```python
## filling the missing value in product_category 3:--
df['Product_Category_3'].unique()
```

```python
df['Product_Category_3'].value_counts()
```

```python
#replacing the missing value with mode because in Product_Category are in discrete variable:--
df['Product_Category_3']=df['Product_Category_3'].fillna(df['Product_Category_3'].mode()[0])
```

```python
df.head()
```

```python
df['Stay_In_Current_City_Years'].unique()
```

```python
# replacing the 4+ with 4  
df['Stay_In_Current_City_Years']=df['Stay_In_Current_City_Years'].str.replace('+','')
```

```python
# After replacing the 4+ with 4:--
df['Stay_In_Current_City_Years'].unique()
```

```python
df.head()
```

```python
df.info()
```

Observation--
      After Checking the info of the data we get that "Stay_In_Current_City_Years" are in (Dtype=object) but the 
      outputs are in numeric value so we have to convert the "Stay_In_Current_City_Years" from object to numeric value.

```python
##convert object into interges:--
df["Stay_In_Current_City_Years"]=df["Stay_In_Current_City_Years"].astype(int)
```

```python
## B & C are also in uint8 so we are going to convert it into int
df['B']=df['B'].astype(int)
df['C']=df['C'].astype(int)
df.info()
```

Note
   1) We have converted "Stay_In_Current_City_Years" from object to integers.    
       2) We have also converted 'B' and 'C' from uint8 to intergers.
